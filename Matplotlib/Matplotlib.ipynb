# Tut 3

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend() #plt.legend("upper right") Default = plt.legend(loc="best")
```

```python
plt.xlabel("Days")
plt.ylabel("Tempreture")
plt.title("Weather")

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")
plt.legend(loc="best",shadow=True, fontsize="small")
plt.grid()
```

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue, label="Revenue")
plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")

plt.bar(company,profit , label="profit", width=0.4)

plt.legend()
```

```python
# barh for horizontal bar graph
plt.barh(company,revenue, label="Revenue")
plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")

plt.barh(company,profit , label="profit")

plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
#By defalut histogram has 10 bins
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=3)
```

```python
# rwidth is relative width with respective bars
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=3,rwidth=0.5)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=3,rwidth=0.95)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color="g")
#Bins
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],rwidth=0.95,color=["green","orange"],label=["Men","Women"])
plt.legend()

plt.xlabel("Sugar Range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
```
